```python
# !pip3 install wordcloud
from datetime import date, timedelta
import twint
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
# import re
from textblob import TextBlob
import nest_asyncio
nest_asyncio.apply()
```

```python
c = twint.Config()

c.Search = ['islamophobia', 'worldcup']       # topic
# c.Since = '2022-10-15'
# c.Until = '2022-12-12'
c.Limit = 1000      # number of Tweets to scrape
# c.Output = "islamophobia_qatar_worldcup.csv"     # path to csv file
c.Store_csv = True       # store tweets in a csv file automatically with name given in output
c.Pandas = True
```

```python
twint.run.Search(c)
```

```python
df = twint.output.panda.Tweets_df
```

```python
df.head()
```

```python
# Convert date colum into a list for the next step
dates_list = df['date'].to_list()
```

```python
# Create a histogram of tweet frequency
# from https://stackoverflow.com/questions/44929555/how-to-properly-create-a-histogram-displaying-the-frequency-of-the-tweets-for-e

dates = []
for t in dates_list:
    # extract the date part of the datetime
    date_str = t.split(' ')[0]
    # extract the time from the date
    year,month,day = [int(i) for i in date_str.split('-')]
    # create a date object
    d = date(year, month, day)
    # sort
    dates.append(d)
    
# sort dates
dates.sort()

# find the first and last date
min_date = dates[0]
max_date = dates[-1]

# compute num days
length = (max_date - min_date).days + 1

# plot histogram
plt.figure(figsize=(12,8))
plt.hist(dates)
plt.show()
```

```python
# Calcuate how many days jack has tweeted during the period collected
df['date_only'] = pd.to_datetime(df['date'])
date_only = df['date_only'].dt.date
num_days = len(date_only.unique())
print("rentang waktu tweet:", num_days)
```

```python
# Create a word cloud for tweets collected, may take sometime
tweets = df['tweet'].to_list()

words = ''
stopwords = set(STOPWORDS)

# Iterate through tweets
for value in tweets:
    # Convert to string
    value = str(value)
    
    # Tokenize
    tokens = value.split()
    # Convert each to lowercase
    for i in range(len(tokens)):
        tokens[i] = tokens[i].lower()
        
    # Add to word
    words += " ".join(tokens)+" "

# Word cloud setup
wordcloud = WordCloud(width = 800, height = 800,
                     background_color = 'white',
                     stopwords = stopwords,
                     min_font_size = 10).generate(words)

# Plot
plt.figure(figsize=(8,8))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

```python
import seaborn as sns
%matplotlib inline

bloblist_tweet = list()

df_tweet=df['tweet'].astype(str)
for row in df_tweet:
    blob = TextBlob(row)
    bloblist_tweet.append((row,blob.sentiment.polarity, blob.sentiment.subjectivity))
    df_tweet_polarity = pd.DataFrame(bloblist_tweet, columns = ['sentence','sentiment','polarity'])
 
def f(df_tweet_polarity):
    if df_tweet_polarity['sentiment'] > 0:
        val = "Positive"
    elif df_tweet_polarity['sentiment'] == 0:
        val = "Neutral"
    else:
        val = "Negative"
    return val

df_tweet_polarity['Sentiment_Type'] = df_tweet_polarity.apply(f, axis=1)

plt.figure(figsize=(10,10))
sns.set_style("whitegrid")
ax = sns.countplot(x="Sentiment_Type", data=df_tweet_polarity)
```
